```python
# create a list of tickers symbol in alphabetical order
tickers=["BAC","C","GS","JPM","MS","WFC"]
```

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
bank_stocks.head()
```

```python
# set the column name levels
bank_stocks.columns.names=['Bank Tickers','Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',level="Stock Info",axis=1).max()
```

```python
# create a new dataframe.this dataframe contains returns for each bank stock.
returns=pd.DataFrame()
returns
```

```python
# We can use pct_change method on close column to create a column representing this return value.
# create a for loop that goes and for each BAnk Stocks tickers
# create this return column.
tickers=["BAC","C","GS","JPM","MS","WFC"]
for tick in tickers:
    returns[tick+"Return"]=bank_stocks[tick]["Close"].pct_change()
```
